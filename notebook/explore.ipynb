```python
from pytorchyolo.models import Darknet
from pytorchyolo.utils.utils import weights_init_normal, non_max_suppression
from pytorchyolo.detect import _create_data_loader, _draw_and_save_output_image, load_classes
import torch
import tqdm
from torch.autograd import Variable
```

```python
model_path = "../config/yolov3.cfg"
device = torch.device("cuda" if torch.cuda.is_available()
                          else "cpu")  # Select device for inference
model = Darknet(model_path).to(device)

model.apply(weights_init_normal)   
model.load_darknet_weights("../weights/yolov3.weights")

model.eval()

scripted_model = torch.jit.trace(model,example_inputs=torch.randn(1,3,416,416).to(device))
scripted_model.save('model_store/yolo_deploy.pt')
```

```python
# Extract class names from file
classes = load_classes('../data/coco.names')  # List of class names
img_path = "../data/samples"
batch_size = 1
img_size = 416
n_cpu = 8
dataloader = _create_data_loader(img_path, batch_size, img_size, n_cpu)
```

```python
conf_thres=0.5
nms_thres=0.5
model.eval()

Tensor = torch.cuda.FloatTensor if torch.cuda.is_available() else torch.FloatTensor
img_detections = []  # Stores detections for each image index
imgs = []  # Stores image paths

for (img_paths, input_imgs) in tqdm.tqdm(dataloader, desc="Detecting"):
        # Configure input
        input_imgs = Variable(input_imgs.type(Tensor))

        # Get detections
        with torch.no_grad():
            detections = model(input_imgs)
            detections = non_max_suppression(detections, conf_thres, nms_thres)
        
        # # Store image and detections
        # img_detections.extend(detections)
        # imgs.extend(img_paths)
        # print(f'Detection for {img_paths}:')

        # _draw_and_save_output_image(img_paths[0],detections,img_size,output_path=".", classes=classes)
```
